```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the CSV file

```python
df = pd.read_csv('poker_dataset.csv')
```

# Extract desired rows and columns

```python
desired_data = df.iloc[:10, :10]
desired_data
```

# Assign indexing to a specific column

```python
df.set_index('hand', inplace=True)
```

# Create a filter depending on 2 parameters and print the appropriate table

```python
filter_condition = (df['result1'] == 'NOTHING') & (df['flop'] == '♥5 ♣2 ♥6') 
filtered_data = df[filter_condition]
print(filtered_data)
```

# Sort the table using the 2 parameters

```python
sorted_data = df.sort_values(by=['result1', 'result2'], ascending=[True, False])
print(sorted_data)
```

# Statistical functions
# Generating random integers between 0 and 100 because my data does, not contains integer columns

```python
new_column_values = np.random.randint(low=0, high=100, size=len(df))
df['new_integer_column'] = new_column_values
specific_column = 'new_integer_column'
mean_value = df[specific_column].mean()
std_deviation = df[specific_column].std()
median_value = df[specific_column].median()
min_value = df[specific_column].min()
max_value = df[specific_column].max()

print("Mean:", mean_value)
print("Standard Deviation:", std_deviation)
print("Median:", median_value)
print("Min:", min_value)
print("Max:", max_value)
```

# Building Chart to visualize frequency of Poker Hand of outcomes

```python
result_counts = pd.concat([df['result1'], df['result2'], df['result3']]).value_counts()

result_counts.plot(kind='bar', figsize=(16, 6))
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Frequency of Poker Hand Outcomes')
plt.xticks(rotation=0)
plt.legend(title='Stage')
plt.show()
```

# Building Result Charts for each one

```python
result_counts = {
    'flop': df['result1'].value_counts(),
    'turn': df['result2'].value_counts(),
    'river': df['result3'].value_counts()
}

fig, axes = plt.subplots(3, 1, figsize=(10, 15))

for i, (stage, counts) in enumerate(result_counts.items()):
    ax = axes[i]
    counts.plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'Results at {stage.capitalize()}')
    ax.set_xlabel('Result')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
